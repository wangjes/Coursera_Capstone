# Data Science Capstone Project - The Battle of Neighborhoods

## Problem

Toronto, the capital of the province of Ontario, is a major Canadian city along Lake Ontario’s northwestern shore. It's a dynamic metropolis with a core of soaring skyscrapers, all dwarfed by the iconic, free-standing CN Tower. In this project, I would like to explore the best location in Toronto to open a Japanese restaurant. Small business owners can be benefitted from this search process to find the best place for them to operate their business.

## Data

To find a solution to our problem, we can make reference to the following data sources:
- Neighborhood information from Wikipedia
- Foursquare API venues explore method
- Python geocoder librarh to get geographical coordinates of the neighborhoods
